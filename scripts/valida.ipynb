# Validación

```python
import geopandas as gpd
import os

directorio_actual = os.getcwd()
directorio_padre = os.path.dirname(directorio_actual)


gdf = gpd.read_file(os.path.join(directorio_padre, "labels", "olaroz_training_polygons.geojson"))

print(gdf.head())
print("CRS:", gdf.crs)

print("Tipos de geometría:")
print(gdf.geom_type.value_counts())

print("Cantidad por clase:")
print(gdf["class_name"].value_counts())

print("Cantidad por split:")
print(gdf["split"].value_counts())

print("Geometrías inválidas:", (~gdf.is_valid).sum())
```

```python
gdf["geometry"] = gdf.geometry.make_valid()
```
